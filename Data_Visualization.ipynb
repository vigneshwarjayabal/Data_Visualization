```python
#Importing necessary modules
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 #using random data points to represent the yield of apples.
yield_apples = [0.895,0.91,0.919,0.926,0.929,0.931]
plt.plot(yield_apples)
```

```python
years = [2010,2011,2012,2013,2014,2015]
yield_apples = [0.895,0.91,0.919,0.926,0.929,0.931]
plt.plot(years,yield_apples)
```

```python
#Let's add labels to the axes so that we can show what each axis represents
plt.plot(years,yield_apples)
plt.xlabel("Year")
plt.ylabel("yield (tons per hectare)");
```

```python
#To plot multiple datasets on the same graph, just use the plt.plot function once for each dataset.
year = range(2000,2012)

apple = [0.895,0.91,0.919,0.926,0.929,0.931,0.934,0.936,0.937,0.9375,0.9371,0.939]
orange = [0.962,0.941,0.930,0.923,0.918,0.908,0.907,0.904,0.901,0.898,0.9,0.896]

plt.plot(year,apple)
plt.plot(year,orange)
plt.xlabel("Year")
plt.ylabel("yield (tons per hectare)");

```

```python
# We can add a legend which tells us what each line in our graph means. To understand what we are plotting, we can add a title to our graph.


plt.plot(year,apple)
plt.plot(year,orange)
plt.xlabel("Year")
plt.ylabel("yield (tons per hectare)");
plt.title("Crop Yield in Kashmir")
plt.legend(['Apples','Oranges'])
```

```python
# To show each data point on our graph, we can highlight them with markers using the marker argument.
# Many different marker shapes like a circle, cross, square, diamond, etc. are provided by Matplotlib.

plt.plot(year,apple,marker='o')
plt.plot(year,orange,marker='x')
plt.xlabel("Year")
plt.ylabel("yield (tons per hectare)");
plt.title("Crop Yield in Kashmir")
plt.legend(['Apples','Oranges'])
```

```python
# You can use the plt.figure function to change the size of the figure.
plt.figure(figsize=(12,6))
plt.plot(year,apple,marker='o')
```

```python
# An easy way to make your charts look beautiful is to use some default styles from the Seaborn library.
# These can be applied globally using the sns.set_style function.

sns.set_style("whitegrid")

plt.plot(year,apple,marker='o')
plt.plot(year,orange,marker='x')
plt.xlabel("Year")
plt.ylabel("yield (tons per hectare)");
plt.title("Crop Yield in Kashmir")
plt.legend(['Apples','Oranges'])

```

```python
sns.set_style("whitegrid")

plt.plot(year,apple,marker='o')
plt.plot(year,orange,marker='x')
plt.xlabel("Year")
plt.ylabel("yield (tons per hectare)");
plt.title("Crop Yield in Kashmir")
plt.legend(['Apples','Oranges'])
```

#**Bar Graph**

image.png

```python
year = range(2000,2006)
apple = [0.35,0.6,0.9,0.8,0.65,0.9]
plt.bar(year,apple)
plt.xlabel("Year")
plt.ylabel("yield (tons per hectare)");
plt.title("Crop Yield in Kashmir")
```

```python
# We can also stack bars on top of each other. Let's plot the data for apples and oranges.
oranges = [0.4,0.8,0.9,0.7,0.6,0.8]
plt.bar(year,apple)
plt.bar(year,oranges, bottom=oranges)
```
