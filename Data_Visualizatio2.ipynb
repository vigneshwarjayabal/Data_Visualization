```python
import pandas as pd
sales = pd.read_csv('/content/drive/MyDrive/train-chennai-sale.csv')
```

```python
sales['AREA'].unique()
```

```python
sales['AREA'] = sales["AREA"].apply(lambda i:"Anna Nagar" if (i == "Ann Nagar" or i == "Ana Nagar") else i)
sales['AREA']=sales["AREA"].apply(lambda i:"Karapakkam" if (i == "Karapakam" ) else i)
sales['AREA']=sales["AREA"].apply(lambda i:"Adyar" if (i == "Adyr") else i)
sales['AREA']=sales["AREA"].apply(lambda i:"Velachery" if (i == "Velchery") else i)
sales['AREA']=sales["AREA"].apply(lambda i:"Chrompet" if (i == "Chrompt" or i == "Chrmpet" or i =="Chormpet") else i)
sales['AREA']=sales["AREA"].apply(lambda i:"KK Nagar" if (i == "KKNagar") else i)
sales['AREA']=sales["AREA"].apply(lambda i:"T Nagar" if (i == "TNagar") else i)

```

```python
sales['AREA'].unique()
```

```python
sales['SALE_COND']=sales["SALE_COND"].apply(lambda i:"AbNormal" if (i == "Ab Normal") else i)
sales['SALE_COND']=sales["SALE_COND"].apply(lambda i:"Partial" if (i == "Partiall" or i == "PartiaLl") else i)
sales['SALE_COND']=sales["SALE_COND"].apply(lambda i:"AdjLand" if (i == "Adj Land") else i)
sales['PARK_FACIL']=sales["PARK_FACIL"].apply(lambda i:"No" if (i == "Noo") else i)
sales['BUILDTYPE']=sales["BUILDTYPE"].apply(lambda i:"Commercial" if (i == "Comercial") else i)
sales['BUILDTYPE']=sales["BUILDTYPE"].apply(lambda i:"Others" if (i == "Other") else i)
sales['UTILITY_AVAIL']=sales["UTILITY_AVAIL"].apply(lambda i:"No sewage" if (i == "NoSewr " or i == "NoSeWa") else i)  #sewage
sales['UTILITY_AVAIL']=sales["UTILITY_AVAIL"].apply(lambda i:"AllPub" if (i == "All Pub") else i)
sales['STREET']=sales['STREET'].apply(lambda i:"Paved" if ( i == "Pavd") else i )
sales['STREET']=sales['STREET'].apply(lambda i:"No Access" if ( i == "NoAccess") else i )
```

```python
sales.isnull().sum()
```

```python
val_col=["QS_ROOMS","QS_BATHROOM","QS_BEDROOM"]
target_col="QS_OVERALL"
def fill_null_QS_OVERALL(row):
  if pd.isnull(row[target_col]):
    return row[val_col].mean()
  else:
    return row[target_col]

sales[target_col]=sales.apply(fill_null_QS_OVERALL,axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# How does the average sales price vary over time?

# You can plot the trend of sales price across different dates (DATE_SALE) to identify trends and patterns over time.
```

```python
#matplotlib
avg_sales_by_date = sales.groupby('DATE_SALE')['SALES_PRICE'].mean().reset_index()
plt.figure(figsize=(10,6))
plt.plot(avg_sales_by_date['DATE_SALE'], avg_sales_by_date['SALES_PRICE'], label='Average Sales Price')
plt.xlabel('Date')
plt.ylabel('Average Sales Price')
plt.title('Average Sales Price Over Time')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
#seaborn
sns.lineplot(x='DATE_SALE', y='SALES_PRICE', data=sales.groupby('DATE_SALE').agg({'SALES_PRICE': 'mean'}).reset_index())
plt.title('Average Sales Price Over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
fig = px.line(avg_sales_by_date, x='DATE_SALE', y='SALES_PRICE', title='Average Sales Price Over Time')
fig.show()
```

```python
# What is the correlation between numerical variables (e.g., SALES_PRICE, INT_SQFT, N_BEDROOM, N_BATHROOM)?

# This question helps visualize the correlation matrix between the numerical columns in the dataset
#  to understand how different features are related to each other.
```

```python
corr_matrix = sales[['SALES_PRICE', 'INT_SQFT', 'N_BEDROOM', 'N_BATHROOM', 'N_ROOM']].corr()
plt.figure(figsize=(10, 6))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=45)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.title('Correlation Heatmap for Numerical Variables')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap for Numerical Variables')
plt.show()
```

```python
fig = px.imshow(corr_matrix, title="Correlation Heatmap for Numerical Variables", color_continuous_scale='RdBu_r')
fig.show()
```

```python
# What is the distribution of properties based on the availability of utility (UTILITY_AVAIL)?

# This question helps visualize the proportion of properties that have access to utility services versus those that do not.
# The donut chart can be particularly effective for showing this kind of breakdown while providing a visually appealing design.
```

```python
utility_avail_counts = sales['UTILITY_AVAIL'].value_counts()

# Matplotlib Donut Chart
plt.figure(figsize=(8, 8))
# Create a pie chart and set the 'wedgeprops' to create a hole in the center (donut effect)
plt.pie(utility_avail_counts, labels=utility_avail_counts.index, autopct='%1.1f%%', startangle=90,
        colors=['lightgreen', 'lightcoral'], wedgeprops={'width': 0.4})
plt.title('Distribution of Properties Based on Utility Availability')
plt.axis('equal')  # Equal aspect ratio ensures the donut chart is circular.
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sns.set_palette('pastel')
plt.pie(utility_avail_counts, labels=utility_avail_counts.index, autopct='%1.1f%%', startangle=90,
        wedgeprops={'width': 0.4})
plt.title('Distribution of Properties Based on Utility Availability')
plt.axis('equal')  # Equal aspect ratio ensures the donut chart is circular.
plt.show()
```

```python
fig = px.pie(sales, names='UTILITY_AVAIL', title='Distribution of Properties Based on Utility Availability', hole=0.4)
fig.show()
```

```python
# Explanation of the Question:
# Distribution of Properties Based on Utility Availability: This donut chart will show how many properties have access to utilities
# like water, gas, or electricity compared to those that do not. This is helpful for understanding
# the percentage of properties with utilities, which may affect property value or desirability.
```

```python
# What is the distribution of properties based on build type (BUILDTYPE) and their respective sale conditions (SALE_COND)?

# This question helps visualize the distribution of properties of different build types and
#  how they are distributed across various sale conditions (e.g., Normal, Abnormal, etc.).
# A stacked bar chart can show the relationship between multiple categorical variables by stacking their proportions.
```

```python
build_sale_counts = pd.crosstab(sales['BUILDTYPE'], sales['SALE_COND'])

# Matplotlib Stacked Bar Chart
build_sale_counts.plot(kind='bar', stacked=True, figsize=(10, 6), color=['skyblue', 'lightgreen', 'salmon', 'orange'])
plt.title('Distribution of Properties Based on Build Type and Sale Condition')
plt.xlabel('Build Type')
plt.ylabel('Count of Properties')
plt.xticks(rotation=45)
plt.legend(title='Sale Condition')
plt.tight_layout()  # Adjust layout to avoid label overlap
plt.show()
```

```python
fig = px.bar(sales, x='BUILDTYPE', color='SALE_COND', title='Distribution of Properties Based on Build Type and Sale Condition',
             category_orders={'SALE_COND': ['Normal', 'Abnormal', 'Partial Sale']},
             labels={'BUILDTYPE': 'Build Type', 'SALE_COND': 'Sale Condition'})
fig.update_layout(barmode='stack')
fig.show()
```
